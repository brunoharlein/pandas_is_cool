# Appliquer des fonctions sur un DataFrame

```python
# DataFrame.apply()
# la methode .apply() effectue la fonction sur chaques colonnes du Dataframe
```

```python
# Soit une fonction qui retourne le 100e élément
def row_100(column):
    # extraire le 100e élément d'une colonne
    item = column.iloc[99]
    return item
```

# Appliquer une fonction à une ligne

```python
# DataFrame.apply(function, axis=1)
```

```python
def is_minior(row):
    if row["age"] < 18:
        return True
    else:
        return False
```

```python
miniors = titanic_survival.apply(is_minior, axis=1)
miniors
```

```python
def which_class(row):
    pclass = row["pclass"]
    # l'element de la colonne pclass de la ligne en question
    if pd.isnull(pclass):
        # Est-ce que cette valeur est manquante ? si oui :
        return "Unknown"
    elif pclass == 1:
        return "First Class"
    elif pclass == 2:
        return "Second Class"
    else :
        return "Third Class"
```

```python
classes = titanic_survival.apply(which_class, axis=1)
classes
```

# Convertir l'âge en une valeur numérique

```python
# Nous allons analyser la colonne age en profondeur. Pour commencer il faut convertir cette colonne en valeur numerique.

# Ce qui nous permettra de trouver l'age moyen des sondés
```

```python
data["Age"].value_counts()
```

- Ecrire une fonction qui converti une chaine de caracteres en une valeur entière. Cela permettra de convertir les valeurs de la colonne "Age" en entiers. Cette fonction prendra en paramètre une chaine de caracteres (les valeurs actuelles de la colonne "Age")


    - Utiliser la fonction isnull() pour vérifier si les valeurs sont nulles. Ajouter une condition if qui retourne "None" si la valeur est nulle.


    - Séparer les chaines de caracteres en fonction de l'espace (" ") et extraire le premier élément de la liste résultante (méthode split())


     - Remplacer le caractere "+" dans le resultat acec une chaine de caracteres vide en remplacement pour le supprimer (méthode replace())
 
 
     - Utiliser int() pour convertir le résultat en entier
     
     
     - Retourner le résultat
     
     
- Utiliser la méthode Series.apply() pour appliquer la fonction à chaque valeurs de la colonne "Age" du DataFrame data.


    - Assigner le resultat à la nouvelle colonne int_age de DataFrame data
    
    
     - Appeler la méthode Series.describe() sur la colonne "int_age" du DataFrame data et afficher le resultat

```python
def extract_age(age_str):
# Prendre la chaine de caracteres de chaque lignes de la colonne "Age"

    if pd.isnull(age_str):
        return None
    
    # Séparation de la chaine de caracteres, l'élément 0 (le premier)
    age_str = age_str.split(" ")[0]
    
    # remplacer le + par du vide 
    age_str = age_str.replace("+", "")
    return int(age_str)

data["int_age"] = data["Age"].apply(extract_age)

data["int_age"].describe()
```

```python
data["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts()
```

- Ecrire une fonction pour convertir les revenus en valeur unique de format entier.

    - Utiliser la fonction isnull() pour vérifier si la valeur est nulle. Si c'est le cas, retourner 'None'

    - Séparer la chaine de caracteres en prenant l'espace comme délimiteur et extraire le premier élément de la liste résultante.

    - Si le résultat vaut "Prefer", retourner "none"

    - Remplacer les caracteres "$" et "," par des chaines de caracteres vides pour les supprimer.

    - Utiliser int() pour convertir le résultat en entier

    - Retourner le résultat

- Utiliser la méthode Series.apply() pour appliquer la fonction précedente à chaque valeur de la colonne "How much total combined money did all members of your HOUSEHOLD earn last year?" du DataFrame data.

    - Assigner le résultat à la nouvelle colonne "int_income" du DataFrame data.

- Appeler la méthode Series.describe() à la colonne int_income du DataFrame data et afficher le résultat.

```python
def extract_income(income_str):
    if pd.isnull(income_str):
        return None
    
    income_str = income_str.split(" ")[0]
    if income_str == "Prefer":
        return None
    
    income_str = income_str.replace(',', '')
    income_str = income_str.replace('$', '')
    return int(income_str)

# création d'une nouvelle colonne 
data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(extract_income)
```
