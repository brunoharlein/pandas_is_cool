# Problème avec les valeurs manquantes

```python
mean_age = sum(titanic_survival["age"]) / len(titanic_survival["age"])
print(mean_age)
```

```python
# Un calcul avec des valeurs manquantes retournera ... des valeurs manquantes !
```

```python
# Il faut donc filtrer les manquants avant de calculer la valeur moyenne
```

Nous allons utiliser age_is_null pour créer un vecteur qui contient seulement les valeurs de la colonne "age" qui ne sont pas de valeur NaN (c'est à dire pour lequelle age_is_null vaut False)

Assigner ce résultat à la variable good_ages

Calculer la moyenne de ce nouveau vecteur et assigner le résultat à la variable mean_age.

Afficher cette moyenne

```python
show_ages = titanic_survival["age"]

print(show_ages)
# Nous affichons les ages de la colonne "age" nous remarquons des NaN
```

```python
age_is_null = pd.isnull(titanic_survival["age"])

print(age_is_null)
# Nous assignons à la variable age_is_null si il y a des valeurs NaN (True)
```

```python
good_ages = titanic_survival["age"][age_is_null == False]

print(good_ages)
# Nous assignons la variable good_ages la colonne age mais seulement le résultats de age_is_null qui valent False
```

```python
# Nous remarquons qu'il reste seulement 1046 résultats dans notre colonne "age"
```

```python
mean_age = sum(good_ages) / len(good_ages)

print(mean_age)
# Pour calculer la moyenne, nous additionnons les ages et nous divisons par le total de personnes.
```

```python
# L'age moyen des survivants du Titanic est de 29.8 
print(sum(good_ages))
print(len(good_ages))
```

```python
# 31255.6667 / 1046 = 29.88113
```

```python
# creation d'un sous ensemble directement dans une colonne et d'effectuer des calculs
# grouper et appliqué des fonctions

# Créez un tableau croisé dynamique de type feuille de calcul en tant que DataFrame.

# DataFrame.pivot_table()
```

# Appliquer des fonctions sur un DataFrame

```python
# DataFrame.apply()
# la methode .apply() effectue la fonction sur chaques colonnes du Dataframe
```

```python
# Soit une fonction qui retourne le 100e élément
def row_100(column):
    # extraire le 100e élément d'une colonne
    item = column.iloc[99]
    return item
```

```python
# Retourner le 100e élément de chaque colonnes
row_100_var = titanic_survival.apply(row_100)
row_100_var
```
