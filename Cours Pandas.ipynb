# Appliquer des fonctions sur un DataFrame

```python
# DataFrame.apply()
# la methode .apply() effectue la fonction sur chaques colonnes du Dataframe
```

```python
# Soit une fonction qui retourne le 100e élément
def row_100(column):
    # extraire le 100e élément d'une colonne
    item = column.iloc[99]
    return item
```

```python
# Retourner le 100e élément de chaque colonnes
row_100_var = titanic_survival.apply(row_100)
row_100_var
```

# Appliquer une fonction à une ligne

```python
# DataFrame.apply(function, axis=1)
```

```python
def is_minior(row):
    if row["age"] < 18:
        return True
    else:
        return False
```

```python
miniors = titanic_survival.apply(is_minior, axis=1)
miniors
```

```python
def which_class(row):
    pclass = row["pclass"]
    # l'element de la colonne pclass de la ligne en question
    if pd.isnull(pclass):
        # Est-ce que cette valeur est manquante ? si oui :
        return "Unknown"
    elif pclass == 1:
        return "First Class"
    elif pclass == 2:
        return "Second Class"
    else :
        return "Third Class"
```

```python
classes = titanic_survival.apply(which_class, axis=1)
classes
```

# Méthode apply() sur les colonnes d'un DataFrame

```python
# Application d'une fonction sur toutes les colonnes du DataFrame
```

```python
# apply()

# Par défaut cela s'appliquera aux colonnes (sauf si vous utilisez l'axis)

# DataFrame.apply(lambda x: x**2) retourne un nouveau dataframe en ayant elevé à la puissance 2 chaque elements des colonnes (evidement des colonnes numeriques)
```

```python
import numpy as np
```

```python
# retourner les types de données en tant que Series

types = fandango_films.dtypes
types
```

```python
# (64) == jusqu'a 64 chiffres apres la virgule
```

```python
# filtrer les nombres à virgules float64 et retourner le nom des colonnes

float_columns = types[types.values == 'float64'].index
float_columns
```

```python
# utilisation des crochets our filtrer les colonnes de la condition float_columns

float_df = fandango_films[float_columns]
float_df
```

```python
# Voici donc un dataframe avec juste les données float64

# Maintenant nous pouvons calculer l'écart type de chaque colonnes (std = standard deviation)

# on applique la std avec la méthode apply sur le DataFrame float_df qui va permettre de calculer l'écart type de chaque colonnes

# x est un objet Series et correspond à chaque colonnes

# Ce qui signifie : applique pour chaque x la fonction std de calcul de l'ecart type

deviations = float_df.apply(lambda x: np.std(x))
print(deviations)
```

```python
# Nous affichons l'écart type pour chaque colonnes
```
